```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn
import time
from pickle import load

from sklearn.preprocessing import StandardScaler,PowerTransformer
import math
from scipy.stats import multivariate_normal
from scipy.stats import norm
import os
```

```python
num_ensemble_members = 40
```

# Lorenz 96 

```python
truth_weather = np.load("Lorenz_96/truth_weather.npy")
array_no_tl = np.load("Lorenz_96/weather_no_tl.npy")
array_tl = np.load("Lorenz_96/weather_tl.npy")
```

```python
expectation_no_tl = np.mean(array_no_tl[:,:,:,:],axis=1)

error_square = np.mean((expectation_no_tl[:,:,:] - truth_weather[:,:,:])**2,axis=(0,1))
    
l96_no_tl_error = error_square**0.5

no_tl_repeat = np.repeat(expectation_no_tl[:,np.newaxis,:,:],num_ensemble_members,axis=1)

spread_square = np.mean((array_no_tl - no_tl_repeat)**2,axis=(0,1,2))
l96_no_tl_spread = spread_square**0.5


expectation_tl = np.mean(array_tl[:,:,:,:],axis=1)

error_square = np.mean((expectation_tl[:,:,:] - truth_weather[:,:,:])**2,axis=(0,1))
    
l96_tl_error = error_square**0.5

tl_repeat = np.repeat(expectation_tl[:,np.newaxis,:,:],num_ensemble_members,axis=1)

spread_square = np.mean((array_tl - tl_repeat)**2,axis=(0,1,2))
l96_tl_spread = spread_square**0.5
```

```python
fig,ax = plt.subplots()

ax.plot(l96_no_tl_error,color="r")
ax.plot(l96_no_tl_spread,linestyle="dashed",color="r")
ax.plot(l96_tl_error,color="#377eb8")
ax.plot(l96_tl_spread,color="#377eb8",linestyle="dashed")
```
